```python
import pandas as pd
import numpy as np

df = pd.read_csv('job_descriptions.csv') 
```

```python
from datetime import datetime, timedelta
```

```python
df.head(4)
```

```python
# Function to generate random timestamps between 10 AM to 1 AM (next day)
def generate_random_timestamp():
    start_time = datetime.strptime('10:00', '%H:%M').time()
    end_time = datetime.strptime('01:00', '%H:%M').time()
    
    # Generate a random number of seconds between 0 and the total seconds in a day
    random_seconds = np.random.randint(0, 24*60*60)
    
    # Calculate the timestamp
    timestamp = (datetime.combine(datetime.today(), start_time) + timedelta(seconds=random_seconds)).time()
    
    return timestamp

# Number of rows in your DataFrame
n = len(df)
```

```python
# Generate timestamps for each row
timestamps = [generate_random_timestamp() for _ in range(n)]

# Add the 'timestamp' column to the DataFrame
df['timestamp'] = timestamps

```

```python

# Step 3: Save the updated DataFrame to a CSV file
df.to_csv('updated_dataset.csv', index=False)

print(f"Updated dataset saved to 'updated_dataset.csv'.")
```
